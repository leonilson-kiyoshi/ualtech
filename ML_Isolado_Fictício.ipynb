```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Gerando dados fictcios
np.random.seed(0)  # Para reproduzibilidade dos resultados

# Mdias de chuvas das duas empresas (cidades)
media_chuvas_cidade1 = 50
media_chuvas_cidade2 = 45  # Mdia menor para cidade com maior volatilidade

# Desvio padro das chuvas das duas empresas (cidades)
desvio_padrao_cidade1 = 5
desvio_padrao_cidade2 = 30  # Maior volatilidade na cidade 2

# Nmero de meses
num_meses = 12

# Gerando dados de chuvas para as duas cidades
chuvas_cidade1 = np.abs(np.random.normal(media_chuvas_cidade1, desvio_padrao_cidade1, num_meses))
chuvas_cidade2 = np.abs(np.random.normal(media_chuvas_cidade2, desvio_padrao_cidade2, num_meses))

# Criando grfico de linhas para mostrar as chuvas das duas cidades
meses = np.arange(1, num_meses+1)

plt.figure(figsize=(10, 5))
plt.plot(meses, chuvas_cidade1, color='blue', marker='o', linestyle='-', label='Cidade 1')
plt.plot(meses, chuvas_cidade2, color='green', marker='o', linestyle='-', label='Cidade 2')
plt.title('Comparao de Chuvas Mensais entre Cidade 1 e Cidade 2')
plt.xlabel('Ms')
plt.ylabel('Chuvas (mm)')
plt.legend()
plt.xticks(meses)
plt.grid(True)
plt.show()

# Calculando mdias das chuvas por cidade
media_cidade1 = np.mean(chuvas_cidade1)
media_cidade2 = np.mean(chuvas_cidade2)

# Ajustando a mdia da cidade com menor volatilidade para ser ligeiramente maior
media_cidade1 += 2

# Criando DataFrame com as mdias de chuvas por cidade
data = {'Cidade': ['Cidade 1', 'Cidade 2'],
        'Mdia de Chuvas (mm)': [media_cidade1, media_cidade2]}
df = pd.DataFrame(data)

print("DataFrame com Mdias de Chuvas por Cidade:")
df
```
